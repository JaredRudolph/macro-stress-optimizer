```python
import pandas as pd
```

```python
df = pd.read_parquet("./data/processed/stress_score.parquet")
```

```python
df = df.dropna()
spy_dd = df["SPY"] / df["SPY"].cummax() - 1
spy_dd_normalized = (spy_dd - spy_dd.min()) / (spy_dd.max() - spy_dd.min())
spy_dd_normalized = 1 - spy_dd_normalized
df["SPY_DRAWDOWN_NORM"] = spy_dd_normalized
```

```python
spy_norm = (df["SPY"] - df["SPY"].min()) / (df["SPY"].max() - df["SPY"].min())
df["SPY_NORM"] = spy_norm
```

```python
df.sort_values(by="Date", ascending=False)
```

```python
df.to_csv("./data/processed/stress_score.csv", index=True)
```
